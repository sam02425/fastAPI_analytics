```python
import requests
path = "/api/events/"
base_url= "http://localhost:8002"
endpoint = f"{base_url}{path}"
endpoint
data = {"page": "/test+"}
import json
json_data = json.dumps(data)
headers = {
    'Content-Type': 'application/json'
}
response = requests.post(endpoint, data=json_data, headers=headers)

print(response.headers)
print("ok",response.ok)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('result'))
else:
    print( response.text)
```

```python
detail_path = "/api/events/1/"
detail_endpoint = f"{base_url}{detail_path}"

response = requests.put(detail_endpoint, json={"description": "inline test"}) # HTTP PUT
print(response.headers)
print('ok', response.status_code)
if response.ok:
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
```

```python
import random
events = 10_000
pages = ['/about', '/contact', '/pages', '/pricing', 'pricing']

path = "/api/events/"
base_url = "http://localhost:8002" # 127.0.0.1
create_endpoint = f"{base_url}{path}"

for i in range(events):
    page = random.choice(pages)
    response = requests.post(create_endpoint, json={"page": page}) # HTTP POST
    if response.ok:
        data = response.json()
        print(data)
```
